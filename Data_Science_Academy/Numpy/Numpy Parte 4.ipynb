### 9. Operações Matemáticas com Matrizes

Uma das operações matemáticas com matrizes é a multiplicação entre elas, aqui ela possui um operador 
específico (@) e segue a regra da álgebra lineae.


```python
import numpy as np
```

```python
# Criação das Matrizes

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
```

```python
# Visualização das Matrizes
print(A)
print()
print(B)
```

Agora vamos executar o produto matricial entre as duas matrizes (diferente da multiplicação de elmento a elemento)

```python
# Produto MATRICIAL

produto_matricial = A @ B
print(produto_matricial)
print()
print(np.dot(A, B))

print()

# Produto NÃO MATRICIAL
print(A * B)
```

Outros exemplos para operações com matrizes

```python
# Definindo matrizes

A = np.array([[10, 20], [30, 40]])
B = np.array([[7, 14], [4, 3]])
```

```python
# Operacoes de Matrizes
soma = A + B
subtracao = A - B
divisao = A / B

print("Operações\n")
print(f"+: {soma}\n")
print(f"-: {subtracao}\n")
print(f"/: {divisao}\n")
```

Os operadores básicos fazem as operações elemento à elemento

```python
# Cortando a Imagem com NumPy

img = Image.open("Imagens/imagem_teste.jpg")

# Conversão de Imagem para Array
array_img = np.array(img)

# Formato do array - três canais R-G-B
# print(array_img.shape)

# Dimensões e canais da imagem
altura, largura, canais = array_img.shape
print(f"Altura: {altura}\nLargura: {largura}\nCanais: {canais}")

# Tamanho do Recorte
recorte_h = 500
recorte_w = 500

# Definindo o centro da Matriz
centro_y = altura // 2
centro_x = largura // 2

# Calculando os limites para o recorte central (operações com matrizes para fatiar a imagem)
x_inicio = centro_x - recorte_h // 2
x_fim = centro_x + recorte_h // 2
y_inicio = centro_y - recorte_w // 2
y_fim = centro_y + recorte_w // 2

# Fazendo o slice do array
corte_central = array_img[y_inicio:y_fim, x_inicio:x_fim]

# Mostra a imagem original e a de corte
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.title("Imagem Original")
plt.imshow(array_img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Corte Central")
plt.imshow(corte_central)
plt.axis("off")

plt.show()
```
