### 2. Comparação do Python Puro com o Numpy

```python
# Criação de uma lista com Numpy de 10 milhões de itens
precos_np = np.random.rand(10_000_000)
type(precos_np)
```

```python
# Criação de uma lista com Python Puro
precos_list = list(precos_np)
type(precos_list)
```

```python
# Medindo o tempo inicial
t0 = time.time()
# Aplicando Operações no vetor np
desc = precos_np * 0.90
final = desc + 5
raiz = np.sqrt(precos_np)
# Exibição do tempo de processamento
print("Numpy: ", time.time() - t0, "segundos")
```

```python
# Medindo o tempo inicial
t0 = time.time()
# Aplicando Operações na lista python
desc = [p * 0.90 for p in precos_list]
final = [p + 5 for p in desc]
raiz = [math.sqrt(p) for p in precos_list]
# Exibição do tempo de processamento
print("Python puro: ", time.time() - t0, "segundos")
```

### 3. Criação de Arrays Numpy

#### 3.1 Criação de um Array - Vetor

```python
# Cria um array numpy a partir de uma lista Python
vector = np.array([17, 21, 100, 34])
print("\nVetor (Array 1D):\n")
print(vector)
```

#### 3.2 Atributos do Array

```python
print("Formato (shape) do vetor: ", vector.shape)
print("Número de dimensões (ndim) do vetor: ", vector.ndim)
print("Número total de elementos (size) do vetor: ", vector.size)
```

#### 3.3 Criação de um Array - Vetor

```python
# Cria um array numpy a partir de uma lista de listas
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("\nMatriz (Array 2D):\n")
print(matriz)
```

#### 3.4 Atributos do Array Matriz

```python
print("Formato (shape) da matriz: ", matriz.shape)             # (linnhas, colunas)
print("Número de dimensões (ndim) da matriz: ", matriz.ndim)
print("Número total de elementos (size) da matriz: ", matriz.size)
```

#### 3.5 Criação de um Array - Tensor

```python
# Cria um array numpu de mais de 2 dimensões (3D, 4D,...)
arr = np.arange(24).reshape(4, 3, 2)
print("\nArray (Array 3D):\n")
print(arr)
```

Aqui o arange recebe a quantidade de elementos do array e o reshape define as dimensõses dele os parâmetros são os eixos X, Y e Z

#### 3.6 Atributos do Tensor

```python
print("Formato (shape) do tensor: ", arr.shape)             
print("Número de dimensões (ndim) do tensor: ", arr.ndim)
print("Número total de elementos (size) do tensor: ", arr.size)
```

Um array com três dimensões vai ter um shape formado por profundidade, linhas e colunas, ou então altura, largura e profundidade em imagens.

#### 3.7 Criação de um Array - Tensor 4D

```python
tensor4d = np.arange(120).reshape(2, 3, 4, 5)
print(tensor4d)
```

#### 3.8 Atributos de um Tensor 4D

```python
print("Formato (shape) do tensor: ", tensor4d.shape)             
print("Número de dimensões (ndim) do tensor: ", tensor4d.ndim)
print("Número total de elementos (size) do tensor: ", tensor4d.size)
```

```python
Aqui nós temos 2x3x4x5 (2 bloos, 3 planos, 4 linhas e 5 colunas)
```
