<h3 style="color: #00ff00;">1. O que é Análise Estatística?</h3>

Análise Estatística é a ciência e a prática de coletar, explorar, interpretar e apresentar dados para descobrir padrões, tendências e relações entre variáveis. Ela utiliza métodos matemáticos e estatísticos para transformar dados brutos em insights compreensíveis.

O objetivo principal da análise estatística não é apenas descrever o que aconteceu no passado (estatística descritiva), mas também usar esses dados para tirar conclusões sobre o mundo em geral, muitas vezes aplicando a teoria da probabilidade para quantificar a incerteza dessas conclusões (estatística inferencial)

A análise estatística fornece as ferramentas para:

- Resumir dados: através de médias, medianas, desvio-padrão, etc;
- Visualizar dados: usando gráficos para identificar padrões;
- Testar hipóteses: para tomar decisões baseadas em evidências (ex: "O novo design do site realmente aumentou as vendas?").
- Quantificar a incerteza: utilizando intervalos de confiança e p-valores para entender a confiabilidade dos resultados

**Principais Tipos de Análise Estatística**
A análise estatística é geralmente dividida em duas grandes categorias:

**1. Estatística Descritiva**
O seu objetivo é resumir e organizar os dados de um estudo ou amostra. Não tenta tirar conclusões para além dos dados analisados. As principais ferramentas incluem:
- Medidas de Tendência Central: Como a média, a mediana e a moda, que representam um valor típico do conjunto de dados.
- Medidas de Dispersão: Como o desvio-padrão, a variância e a amplitude, que indicam o quão espelhados os dados estão.
- Visualizações Gráficas: Gráficos de barras, histograas e diagramas de dispersão que ajudam a visualizar os dados

**2. Estatística Inferencial**
O eu objetivo é usar os dados de uma amostra para fazer generalizações, previsões ou inferências sobre uma população inteira da qual a amostra foi retirada. As técnicas mais comuns incluem:
- Testes de Hipóteses: Procedimentos formais usados para aceitar ou rejeitar uma hipótese estatística (ex: testar se um novo medicamento é mais eficaz que o antigo);
- Análise de Regressão: Modela a relação entre uma variável dependente e uma ou mais variáveis independentes, permitindo fazer previsões;
- Intervalos de Confiança: Fornecem um intervalo de valores que provavelmente contém o verdadeiro valor de um parâmetro da população.

<h3 style="color: #00ff00;">4. Diferença entre Modelagem Estatística e Modelagem Preditiva</h3>

**1. Modelagem Estatística**

O objetivo pricipal é explicar fenômenos. Ela busca compreender como e por que as variáveis estão relacionadas. Por exemplo, em uma regressão linear clássica, o foco é estimar o impacto de cada variável independente sobre a variável dependente, interpretando coeficientes, intervalos de confiança e significância estatística (p-valores). Aqui, a explicabilidade é mais importante do que a precisão. É uma abordagem comum em pesquisas científicas, economia e ciências sociais.

**2. Modelagem Preditiva**

A modelagem preditiva é orientada à performace e generalização. O interesse não está necessariamente em entender o "porquê", mas em fazer boas previsões, seja para identificar clientes propensos ao churn, prever fraudes ou recomendar produtos. Ela usa algoritmos mais complexos (ex: árvores de decisão, random forests, gradient boosting, redes neurais e entre outros). Esss técnicas se apoiam em ajustes finos (ex: cross-validation, otimização de hiperparâmetro). Aqui, a prioridade é o acerto, mesmo que o modelo seja uma "caixa-preta".

<h4 style="color: #00ff00;">5.1 Estatística Descritiva</h4>

É o processo de usar números para resumir e descrever as principais características de um conjunto de dados. Em vez de olhar para centenas de dados brutos, usamos medias como a média (valor central), mediana (valor do meio) e desvio padrão (o quão espelhados os dados estão).

<h5 style="color: #00ff00;">5.1.1 Medidas de Tendência Central</h5>

As medidas de tendência central indicam onde os dados se concentram, ou seja, o valor que representa o "centro" ou o "típico" do conjunto. As principais são:

- Média (aritmética): É uma soma de todos os valores divididos pelo número de observações. Representa o valor médio do grupo, mas é sensível a outliers (valores muito altos ou baixos). Ex: a média de (2, 3, 5) é (2+3+5)/3 = 3.33
  
- Mediana: É o valor central quando os dados estão ordenados. Divide o conjunto ao meio, metade dos valores ficam abaixo e metade dos valores ficam acima. É menos afetado por outliers, sendo útil em distribuições aritméticas. Ex: mediana de (2, 3, 100) é 3.

- Moda: É o valor que mais se repete no conjunto. Pode haver uma (unimodal), várias (multimodal) ou nenhuma moda (quando todos os valores são únicos). Ex: a moda de (2, 2, 3, 5) é 2, pois ele é o valor que se repete.


Essas medidas respondem à pergunta: "Qual o valor típico ou central dos meus dados?"

<h5 style="color: #00ff00;">5.1.2 Medidas de Dispersão</h5>

As medidas de dispersão mostram o quanto os dados se afastam do centro, ou seja, o grau de variabilidade entre as observações. As principais são:

- Amplitude (ou intervalo): É a diferença entre o maior e o menor valor. Mede a variação total, mas depende fortemente de outliers. Ex: se os dados vão de 10 até 40 a amplitude é de 30.

- Variância: Mede quanto, em média, os valores se afastam da média, elevando essas diferenças ao quadrado. É expressa em unidades ao quadrado. Ex:

- Desvio Padrão: É a raiz quadrada da variãncia, voltando à unidade original dos dados. Representa o desvio médio em relação à medida. Ex: se o desvio padrão é 5, os dados tendem a variar ± 5 unidade em torno da média.

- Coeficiente de Variação (CV): É o desvio padrão dividido pela média, em percentual. Mostra a dispersão relativa - quanto maior o CV, mais instável é o conjunto de dados. Ex: CV = (DP/Media) = 100.

*Resumindo*
- As medidas de tendência central indicam onde os dados se concentram;
- As medidas de dispersão indicam o quanto eles variam em torno desse centro.

<h4 style="color: #00ff00;">5.2 Análise Descritiva</h4>

```python
# Para resultados reprodutíveis
np.random.seed(42)

# Cria dados aleatórios seguindo a distribuição normal
group_a = np.random.normal(loc = 175, scale = 5, size = 100) # Média ~175cm e DP ~5
group_b = np.random.normal(loc = 180, scale = 6, size = 100) # Média ~189cm e DP ~6
```

```python
# Cria e mostra o DataFrame
df_alturas = pd.DataFrame({"Grupo A": group_a, "Grupo B": group_b})
df_alturas.head()
```

```python
print("------------- ESTATÍSTICAS DESCRITIVAS COM PANDAS -------------")
df_alturas.describe()
```

```python
# Estatísticas com Statsmodels
dsc_statsmodels_a = sm.stats.DescrStatsW(group_a) # Estatística para o grupo A
dsc_statsmodels_b = sm.stats.DescrStatsW(group_b) # Estatística para o grupo B

print(f"Média: {dsc_statsmodels_a.mean:.2f}")
print(f"DP: {dsc_statsmodels_a.std:.2f}")
print(f"Min: {np.min(group_a):.2f}")
print(f"Max: {np.max(group_a):.2f}")

print(f"\nMédia: {dsc_statsmodels_b.mean:.2f}")
print(f"DP: {dsc_statsmodels_b.std:.2f}")
print(f"Min: {np.min(group_b):.2f}")
print(f"Max: {np.max(group_b):.2f}")
```

Para os valores mínimos e máximos usamos o NumPy, pois esse tipo de estatística é muito simples e pode ser obtida com ferramentas estudadas anteriormente. O statsmodels trabalha com estatísticas mais complexas.

```python
# Visualização em Gráfico de uma distribuição Normal
sns.kdeplot(data=df_alturas, fill=True)
plt.title("Distribuição Normal de Alturas dos Grupos A e B")
plt.xlabel("Altura (cm)")
plt.ylabel("Densidade")
plt.show()
```

<h5 style="color: #00ff00;">5.2.1 Análise</h5>

1. Tendência Central (Média)
    - Grupo A: média de 174,48 cm
    - Grupo B: média de 180,13 cm
    
        Isso mostra que, em média, o Grupo B é mais alto que o Grupo A, cerca de 5,6 cm a mais. Essa diferença é considerável, especialmente se estivessemos analisando população humana ou amostras que deveriam ser semelhantes.
    
        Em uma análise estatística, essa diferença na média poderia ser testada com um teste t de médias independentes, para verificar se a diferença é estatisticamente significativa (ou seja, não causada apenas por acaso).

2. Dispersão (Desvio Padrão)
    - Grupo A: desvio padrão de 4,52 cm

        Em termos práticos: A altura das pessoas no Grupo A tende a se concentrar mais em torno da média, enquanto no Grupo B há mais variabilidade.

4. Conclusão

   O Grupo B é, em média, mais alto e mais heterogêneo do que o Grupo A. Isso significa que, além de possuir uma média superior, o Grupo B apresenta maior variabilidade de alturas, possivelmente refletindo uma amosta mais diversificada ou influenciada por fatores adicionais (por ex: idade, origem ou gênero, dependendo do contexto).

<h4 style="color: #00ff00;">5.3 Análise de Correlação em Python</h4>

**Definição:** Correlação é uma medida estatística que expressa o grau em que duas variáveis se movem em relação uma à outra.

- **Correlação Positiva:** Quando uma variável aumenta, a outra tende a aumentar (ex: altura e peso);
- **Correlação Negativa:** Quando uma variável aumenta, a outra tende a diminuir (ex: horas de estudo e horas de videogame);
- **Correlação Próxima de Zero:** Não há uma relação linear clara entre as variáveis

O coeficiente de correlação varia de -1 a +1

A correlação é, em essência, uma medida da estatística descritiva, pois descreve a força e a direção da relação entre duas variáveis dentro de um conjunto de dados observado.

O coeficiente de correlação (como o de Pearson ou Spearman) é calculado diretamente a partir dos dados de amostra e resume essa relação numericamente. Por exemplo, um valor próximo de +1 indica uma relação positiva forte, enquanto um valor próximo de -1 indica uma relação negativa forte.

```python
np.random.seed(101)

# Gerando dados: Anos e Experiência vs. Salário
anos_exp = np.arange(1, 11) + np.random.randn(10) * 0.5
salario = 5000 + anos_exp * 1500 + np.random.randn(10) * 800
df_salario = pd.DataFrame({"Anos de Experiência": anos_exp, "Salário": salario})
```

```python
df_salario.head()
```

```python
# Calculando o coeficiente de correlação de Pearson
correlacao = df_salario["Anos de Experiência"].corr(df_salario["Salário"])
print(f"\n---Coeficiente de Correlação---")
print(f"A correlação entre Anos de Experiência e Salário é: {correlacao:.4f}\n")
```

```python
# Visualizando a correlação com um gráfico de dispersão
sns.regplot(x = "Anos de Experiência", y = "Salário", data=df_salario)
plt.title("Relação entre Anos de Experiência e Salário")
plt.xlabel("Anos de Experiência")
plt.ylabel("Salário (R$)")
plt.show()
```

```python
# Calcula a Matriz de Correlação entre os dois grupos
correlacoes = df_alturas.corr()

# Exibe o heatmap
plt.figure(figsize=(6,4))
sns.heatmap(correlacoes, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap de Correlações entre os grupos")
plt.show()
```

<h3 style="color: #00ff00;">6. Estatística Inferencial</h3>

**Descrição:** Estatística Inferencial é o ramo da estatística que permite tirar conclusões ou fazer inferências sobre uma população com base em uma amostra de dados. Em vez de apenas descrever os dados observados (como na estatística descritiva), a estatística inferencial utiliza métodos probabilísticos para estimar parâmetros populacionais, testar hipóteses e avaliar a significância de resultados.

A estatística inferencial busca responder perguntas como:

"O que podemos inferir sobre toda a população, a partir desta amostra?"

Ferramentas típicas incluem intervalos de confiança, testes de hipóteses, análises de variância (ANOVA) e regressões, sempre com o objetivo de transformar dados amostrais em conhecimento generalizável e confiável.

<h4 style="color: #00ff00;">6.1 Testes de Hipóteses (Teste t)</h4>

**Definição:** É um método formal para tomar uma decisão sobre uma afirmação (hipótese) usando dados. O **Teste t**, especificamente, é usado para comparar as médias de dois grupos e determinar se a diferença entre elas é estatísticamente significativa ou se provavelmente ocorreu por acaso.

- **Hipótese Nula (H<sub>0</sub>):** A afirmação padrão, geralmente de que "não há diferença". (Ex: "A média de altura dos dois grupos é a mesma");
- **Hipótese Alternativa (H<sub>1</sub> ou H<sub>a</sub>):** É a afirmação contrária à hipótese nula (H<sub>0</sub>), ela representa o que queremos comprovar ou evidenciar com os dados.
- **Valor-p (p-value):** A probabilidade de observar os dados que temos se a Hipótese Nula for verdadeira. Um valor-p baixo (geralmente < 0.05) nos dá evidências para rejeitar a Hipótese Nula.

```python
# Vamos usar os mesmos dados de altura dos grupos A e B
# Hipótese Nula - Não há diferença significativa entre as médias de altura do Grupo A e do Grupo B
# Hipótese Nula - Não há diferença entre as médias dos grupos.

# Hipótese Alternativa - Há diferenças entre as médias dos grupos.

# Unicaudal e Bicaudal

# Realizando o Teste t com Statsmodels
# ttest_ind retorna: (estatistica t, valor-p, graus de liberdade)
t_stat, p_value, df = ttest_ind(group_a, group_b)

print(f"\n---Resultado do Teste t---\n")
print(f"Estatística t: {t_stat:.4f}")
print(f"Valor-p: {p_value:.4f}")

# Interpretando o resultado
alpha = 0.05
if p_value < alpha:
    print(f"\nConclusão: Como o valor-p {p_value:.4f} é menor que {alpha}, rejeitamos a Hipótese Nula.")
    print("Há uma diferença estatisticamente significativa entre as médias dos dois grupos.")
else:
    print(f"\nConclusão: Como o valor-p ({p_value:.4f}) é maior que {alpha}, não temos evidências para rejeitar a Hipótese Nula.")
    print("Não há uma diferença estatisticamente significativa entre as médias dos dois grupos.")
```

<h4 style="color: #00ff00;">6.2 Regressão Linear Simples</h4>

**Definição:** É uma técnica para modelar a relação entre uma variável independente (preditora, ``X``) e uma variável dependente (resposta, ``Y``) através de uma linha reta. O objetivo é não apenas prever ``Y`` a partir de ``X``, mas também entender a força e a direção dessa relação.

A equação é: $Y = β$<sub>0</sub>$+β$<sub>1</sub>$X + ε$

- $β$<sub>0</sub>: Intercepto (onde a linha cruza o eixo Y)
- $β$<sub>1</sub>: Coeficiente Angular (quanto ``Y`` muda para cada unidade de aumento em ``X``)

Verifique o manual em PDF sobre Regressão Linear Múltipla para mais informações.

```python
# 1. Criação de dados de exemplo
# Ex: estudando a relação entre horas de estudo e nota em um exame

np.random.seed(42)
horas_estudo = np.random.uniform(1, 10, 50) # Variável independente X
nota = 5 + 3 * horas_estudo + np.random.normal(0, 3, 50) # Variável dependente Y
```

```python
# Cria o DataFrame
df_ = pd.DataFrame({"Horas_Estudo": horas_estudo, "Nota": nota})
df_.head()
```

```python
# 2. Visualização dos Dados
sns.scatterplot(data = df_, x = "Horas_Estudo", y = "Nota")
plt.title("Relação entre Horas de Estudo e Nota")
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota")
plt.show()
```

```python
# 3. Ajuste do Modelo de Regressão Linear
X = sm.add_constant(df_["Horas_Estudo"]) # Adiciona o Intercepto, o que é requerido pelo Statsmodels (uma constante)
y = df_["Nota"]
```

```python
# OLS = Ordinary Least Squares (Minimos Quadrados Ordinários)
modelo_df_ = sm.OLS(y, X).fit()
```

```python
# 4. Resultados do modelo
print(modelo_df_.summary())
```

Os erros padrão (Standard Errors) apresentados assumem que a matriz de covariância dos erros está corretamente especificada. 

Em outras palavras. O modelo de regressão está supondo que os erros (resíduos) seguem as condições ideais do modelo clássico de mínimos quadrados (OLS). Isso significa que o Statsmodels calculou os erros padrão, os valores-t e os p-valores sob as seguintes subposições:

- Homoscedasticidade: a variância dos erros é constante ao longo das observações (não há heterocedasticidade);
- Independência do erros: os resíuos não estão correlacionados entre si;
- Especialização correta: o modelo inclui todas as variáveis relevantes e não há omissão de termos importantes.

Se alguma dessas condições for violada, então os erros padrão calculados podem estar incorretos - e, consequentemente, os valores-t e os p-valores podem não ser confiáveis para testar hipóteses sobre os coeficientes.

```python
# Extraindo os coeficientes do modelo (o que o modelo aprendeu durante o treinamento)
b0 = modelo_df_.params["const"]
b1 = modelo_df_.params["Horas_Estudo"]
```

```python
# Cria o gráfico
sns.regplot(data=df_,
            x = "Horas_Estudo",
            y = "Nota",
            ci = None,
            line_kws = {"color": "green"}
           )

# Título e rótulos
plt.title("Regressão Linear: Horas de Estudo x Nota no Exame")
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota no Exame")

# Adiciona a fórmula e o R² no gráfico
plt.text(x = df_["Horas_Estudo"].min(),
         y = df_["Nota"].max() - 2,
         s = f'y = {b0:.2f} + {b1:.2f}x\nR² = {modelo_df_.rsquared:.2f}',
         fontsize = 12,
         color = 'black',
         bbox = dict(facecolor="yellow", alpha=0.7, edgecolor="gray"))
plt.show()
```

**Nota: Todo o modelo é uma aproximação e isso é suficiente para fazer inferências e análises (de acordo com a performace do modelo), sempre sabendo que há uma margem de erro. Isso vale para qualquer modelo de Inteligência Artificial na Terra.**
